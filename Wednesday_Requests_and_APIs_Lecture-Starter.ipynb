### Python Wrappers

A lot of APIs have Python 'wrappers' or development kits that will make it easier to find data you're searching for. We're going to make a simple version of our own to see how that works...

```python
# for each request on the weather API, you'll need an HTTP address:
# build it with the information the API says you need:

# base url
base_url = 'http://api.weatherapi.com/v1'
# API method (in this case, you can get current, forecast, history, etc)
api_method = '/current.json'
# API key, which authenticates (signed in as authenticated user)
api_key = '498136789bd44428b3e185934221805'
# q is the query you want...
q = 'Oakland'


# now build the URL
# ?= makes query/search
# &= adds another thing to the query/search

request_url = f"{base_url}{api_method}?key={api_key}&q={q}"
print(request_url)
```

```python
weather_res = requests.get(request_url)
weather_res
```

```python
weather_data = weather_res.json()
```

```python
city = weather_data['location']['name']
state = weather_data['location']['region']
current_temp = weather_data['current']['temp_f']
current_condition = weather_data['current']['condition']['text']
feels_like = weather_data['current']['feelslike_f']
```

```python
# This is all fine and dandy, but it would be better to 
# create an object (class) that then has methods to call different things...
```

```python
class CurrentWeather:
    def __init__(self, city, state, temp, condition, feels_like):
        self.city = city
        self.state = state
        self.temp = temp
        self.condition = condition
        self.feels_like = feels_like
        
    def __str__(self):
        degree_sign = u'\N{DEGREE SIGN}'
        return f"{self.city}, {self.state} - {self.condition.title()} - {self.temp}{degree_sign}F"
    
    def __repr__(self):
        degree_sign = u'\N{DEGREE SIGN}'
        return f"<Weather {self.city} | {self.temp} {degree_sign}F>"
```

```python
weather = CurrentWeather(city, state, current_temp, current_condition, feels_like)
print(weather)
print(weather.condition)
```

```python
# But this is still not a wrapper... 
# we want to create a function or object-method that will allow us to input ANY city and return its info

# so the object is the API or database...
class WeatherAPI:
    
    # we'll input info to access the API
    # if you don't hard-code in your specific API key, others can copy this code and use it in their own stuff
    def __init__(self, api_key):
        self.api_key = api_key
        self.base_url = 'http://api.weatherapi.com/v1'
        
    # now set up '.get requests' based on what information user is searching for
    # when you put an underscore before the method name,
    # it signals to other coders that this is specific to the writer of the code 
    # and not available to others...
    
    # pass in a q to know what info to get, and an API method, which tells which part of the API to query
    def _get(self, q, api_method):
        request_url = f"{self.base_url}{api_method}?key={self.api_key}&q={q}"
        weather_response = requests.get(request_url)
        if weather_response.ok:
            return weather_response.json()
        else:
            print("There was an error...")

    def get_current_weather(self, city):
        weather_data = self._get(city, '/current.json')
        city = weather_data['location']['name']
        state = weather_data['location']['region']
        current_temp = weather_data['current']['temp_f']
        current_condition = weather_data['current']['condition']['text']
        current_feels_like = weather_data['current']['feelslike_f']
        current_weather = CurrentWeather(city, state, current_temp, current_condition, current_feels_like)
        return current_weather
        
```

```python
# now that we have a 'wrapper'
# we can write a little program that takes user input

def main():
    api_key = input('What is your API key? ')
    client = WeatherAPI(api_key)
    while True:
        city = input("Type a city to get its current weather conditions... type \'quit\' to exit ")
        if city.lower() == 'quit':
            break
        method = input("Are you looking for the current or forecasted weather? ")
        while method not in {'current', 'forecast'}:
            method = input("Please type 'current' or 'forecast' ")
        if method == 'current':
            city_weather = client.get_current_weather(city)
            print(city_weather)
        else:
            days = input("For how many days in the future are you interested (limit 3) ")
            client.get_forecast(city, days)
            
main()
```

```python
client = WeatherAPI(api_key)

chicago_weather = client.get_current_weather('Chicago')
print(chicago_weather)

sf_weather = client.get_current_weather('San Francisco')
print(sf_weather)
```

#### In-class Exercise

Add a method for getting the forecast weather for a city. It should default to 3 days in the future. It should display the high, low and condition with the date.

```python
class WeatherAPI:
    
    def __init__(self, api_key):
        self.api_key = api_key
        self.base_url = 'http://api.weatherapi.com/v1'
        
    # set up to input any number of other inputs...
    # that way we can add other things to get, after the required elements in the get request
    def _get(self, q, api_method, **kwargs):
        request_url = f"{self.base_url}{api_method}?key={self.api_key}&q={q}"
        # to set up a way to add pieces onto the base url, 
        # ask Python to 'get' a keyword; if there isn't anything, Python returns NONE
        days = kwargs.get('days')
        # now test for whether there is a 'days' kwarg:
        if days:
            request_url += f"&days={days}"
        weather_response = requests.get(request_url)
        if weather_response.ok:
            return weather_response.json()
        else:
            print("There was an error...")

    def get_current_weather(self, city):
        weather_data = self._get(city, '/current.json')
        city = weather_data['location']['name']
        state = weather_data['location']['region']
        current_temp = weather_data['current']['temp_f']
        current_condition = weather_data['current']['condition']['text']
        current_feels_like = weather_data['current']['feelslike_f']
        city_current_weather = CurrentWeather(city, state, current_temp, current_condition, feels_like)
        return city_current_weather
        
    # API method for forecast also requires a days input
    def get_forecast(self, city, days=3):
        weather_data = self._get(city, '/forecast.json', days=days)
        degree_sign = u'\N{DEGREE SIGN}'
        state = weather_data['location']['region']
        list_of_days = weather_data['forecast']['forecastday']
        print(f"Forecast for {city}, {state}: ")
        for day in list_of_days:
            date = day['date']
            high = day['day']['maxtemp_f']
            low = day['day']['mintemp_f']
            condition = day['day']['condition']['text']
            print(f"On {date}, it will be {condition} with a high of {high}{degree_sign}F and a low of {low}{degree_sign}")

main()
```

#### Pokemon API Wrapper

Create a python wrapper for the Pokemon API. It should take in a pokemon name and display the pokemon with its height and weight

```python
# sets up a formatted output for selected Pokemon

class Pokemon:
    
    def __init__(self, name, height, weight):
        self.name = name
        self.height = height
        self.weight = weight
        
    def __str__(self):
        return f"\n{self.name.title()} is {self.height} high and weighs {self.weight}.\n"
    
    def __repr__(self):
        return f"<Pokemon: {self.name} | height: {self.height} | weight: {self.weight}>"
        
```

```python
# no API key
# make a 'main()' function to allow user-input for name


class PokemonInfo:
    # base url -
    #     https://pokeapi.co/api/v2/pokemon/{id or name}/
    def __init__(self):
        self.base_url = 'https://pokeapi.co/api/v2/pokemon/'
    
    # make request for JSON response (API dictionary)
    def _get(self, name):
        request_url = f"{self.base_url}{name}"
        poke_response = requests.get(request_url)
        if poke_response.ok:
            return poke_response.json()
        else:
            print("There was an error...")

    def get_poke_info(self, name):
        poke_info = self._get(name)
        height = poke_info['height']
        weight = poke_info['weight']
        poke_describe = Pokemon(name, height, weight)
        return poke_describe
```

```python
def main_poke():
    name = input("What Pokemon would you like to learn about? ")
    try: 
        poke_info = PokemonInfo()
        my_poke = poke_info.get_poke_info(name)
        print(my_poke)
    except:
        print("Sorry, we couldn't find that Pokemon.")
    
main_poke()
```
